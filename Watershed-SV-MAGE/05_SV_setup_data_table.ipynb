This notebook sets up a new Terra data table (`watershed_sv_run`) where we store the inputs and outputs of the Watershed-SV workflow.

This notebook should take about 1 minute to run.

## Setup

```r
BiocManager::install('AnVILGCP')
```

```r
library(AnVILGCP)
library(data.table)
```

Similar to notebook 1, the `avtable()` function is used to import Terra data tables into R as a data frame.

```r
anvil_file_tbl <- setDT(avtable('anvil_file'   ))
anno_meta_tbl  <- setDT(avtable('anno_metadata'))
```

## Construct Terra data table specifying workflow inputs

Here we pull file references from the `anvil_file` and `anno_metadata` tables and organize them into a new, single-row Terra data table. Each column of this new table represents an input to the Watershed-SV annotation workflow.

```r
watershed_sv_run_tbl <-
  cbind(
    anvil_file_tbl[, .(fnm=`pfb:file_name`, `pfb:drs_uri`) |> transpose(make.names='fnm')],
    anno_meta_tbl [, .(fnm=basename(file),       file    ) |> transpose(make.names='fnm')]
  )[, .(# Renaming and keeping only the needed files.
        # The name of the first column determines the name of the Terra table.
        watershed_sv_run = 'MAGE_1000G',
    
        # Annotation data
        ABC_enhancers        = ABC_enhancers.merged.All.sorted.11col.region.hg38.bed,
        bw_CADD              = `CADD_GRCh38-v1.6.bw`,
        bw_GC                = gc5Base.bw,
        bw_linsight          = LINSIGHT_hg38.bw,
        bw_phastcon          = hg38.phastCons20way.bw,
        cpg_file             = cpgIslandExt.txt,
        enhancers            = matrix_hs.csv,
        gencode_genes        = gencode.v26.GRCh38.genes.gtf,
        genome_bound_file    = gtex_human_reference.genome,
        primary_cells        = hs_primary.txt,
        remap_crm            = remap2020_crm_macs2_hg38_v1_0.bed.gz,
        roadmap_tar          = roadmap_all_gtex.tar.gz,
        TADs_tar             = TAD.tar.gz,
        vep_cache_tar        = vep_cache.tar.gz,
        
        # Genetic/phenotypic data
        metadata        = sample.metadata.MAGE.v1.0.txt,
        expression_file = paste0(avstorage(),'/data/derived/mage_expression-watershed_sv_input.tsv'),
        gene_list       = paste0(avstorage(),'/data/derived/mage_gene_list.csv'),
        input_vcf       = paste0(avstorage(),'/data/derived/ALL.wgs.mergedSV.v8.20130502.svs.genotypes-edited.vcf.gz'),
        maf_file        = paste0(avstorage(),'/data/derived/1kGp3_sv_maf_file.tsv'),
    
        # Watershed-SV options
        CN_mode = 'extract',
        collapse_instruction = collapse_annotation_instructions.tsv,
        collapse_mode = 'gene-sv',
        expression_field = 'MedZ',
        exprssion_id_field = 'Ind',
        filter_ethnicity = 'false',
        filter_rare = 'false',
        filters = '[".", "PASS"]',
        flank = 1e4,
        length_field = 'SVLEN',
        length_mode = 'extract',
        maf_field = 'Max_MAF',
        maf_mode = 'upload',
        min_support_tissue = 1,
        pipeline = 'population',
        remove_control_genes = 'true',
        zscore_threshold = 3
      )
]
watershed_sv_run_tbl
```

## Upload
`AnVILGCP::avtable_import()` is used to upload an R dataframe as a Terra data table.

```r
avtable_import(watershed_sv_run_tbl, 'watershed_sv_run')
```
