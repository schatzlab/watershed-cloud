Here, the gene-individual pair list of outliers created in notebook 3 (`mage_expression-watershed_sv_input.tsv`) and the gene-individual pair annotations from the workflow (`combined_dataset.tsv`) are subset to the unique individuals & genes in the predetermined `400_random_pairs.csv` file.

```r
dt1 <- fread('mage_expression-watershed_sv_input.tsv')[, gene := sub('\\..*','',gene    )]
dt2 <- fread('combined_dataset.tsv'                  )[, gene := sub(  ':.*','',GeneName)]

samps <- fread('400_random_pairs.csv')$SubjectID
genes <- fread('400_random_pairs.csv')$GeneName |> sub(pat=':.*', rep='')

dt1_400 <- dt1[Ind       %in% samps & gene %in% genes]
dt2_400 <- dt2[SubjectID %in% samps & gene %in% genes]

fwrite(dt1_400, 'mage_expression-watershed_sv_input-400_random_pairs.tsv', sep='\t')
fwrite(dt2_400, 'combined_dataset-400_random_pairs.tsv',                   sep='\t')

system('gcloud storage cp combined_dataset-400_random_pairs.tsv $WORKSPACE_BUCKET/data/derived/')
system('gcloud storage cp mage_expression-watershed_sv_input-400_random_pairs.tsv $WORKSPACE_BUCKET/data/derived/')
```
